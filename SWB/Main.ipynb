```python
import re
import time
import zipfile
import calendar
import numpy as np
import os

year_start = 1998
year_end = 2019
month = 0
year = 0

months = np.zeros(((year_end-year_start+1)*12,2), dtype=int)

y = year_start
m = 1
for i in range (0,(year_end-year_start+1)*12):
    months[i,0] = m
    months[i,1] = y
    if m>11:
        m = 1
        y +=1
    else:
        m += 1

file_name_tmin = "D:\Andrea\SWB_Po\precip\\tmin_%04d.zip" % (year_start)
z_tmin = zipfile.ZipFile(file_name_tmin)
file_name_tmax = "D:\Andrea\SWB_Po\precip\\tmax_%04d.zip" % (year_start)
z_tmax = zipfile.ZipFile(file_name_tmax)
file_name_precip = "D:\Andrea\SWB_Po\precip\\precip_%04d.zip" % (year_start)
z_precip = zipfile.ZipFile(file_name_precip)
i = 0
numberOfDays = calendar.monthrange(months[i,1], months[i,0])[1]
for day in  list(range(1, numberOfDays+1)):
    day_file_tmin = "tmin_%04d_%02d_%02d.asc" % (months[i,1],months[i,0],day)
    z_tmin.extract(day_file_tmin, path="D:\Andrea\SWB_Po\precip")
    day_file_tmax = "tmax_%04d_%02d_%02d.asc" % (months[i,1],months[i,0],day)
    z_tmax.extract(day_file_tmax, path="D:\Andrea\SWB_Po\precip")
    day_file_precip = "precip_%04d_%02d_%02d.asc" % (months[i,1],months[i,0],day)
    z_precip.extract(day_file_precip, path="D:\Andrea\SWB_Po\precip")
ran = ((year_end-year_start+1)*12)

while (i < ran):
    for wtim in range(1,1000000): #138 h for a sleep of 0.5 s
        with open('D:\Andrea\SWB_Po\\display.txt') as f:
            str = f.read()
            last_char = str[-24:]
            date = re.findall("\d+", last_char)
            if len(date) >5:
                month = int(date[0]+date[1])
                year = int(date[-4]+date[-3]+date[-2]+date[-1])        
                if months[i,0] == month:
                    i += 1            
                    numberOfDays = calendar.monthrange(months[i,1], months[i,0])[1]
                    for day in  list(range(1, numberOfDays+1)):
                        day_file_tmin = "tmin_%04d_%02d_%02d.asc" % (months[i,1],months[i,0],day)
                        z_tmin.extract(day_file_tmin, path="D:\Andrea\SWB_Po\precip")
                        day_file_tmax = "tmax_%04d_%02d_%02d.asc" % (months[i,1],months[i,0],day)
                        z_tmax.extract(day_file_tmax, path="D:\Andrea\SWB_Po\precip")
                        day_file_precip = "precip_%04d_%02d_%02d.asc" % (months[i,1],months[i,0],day)
                        z_precip.extract(day_file_precip, path="D:\Andrea\SWB_Po\precip")
                    mexage = 'exctracted all days of %02d/%04d'  % (months[i,0],months[i,1]) 
                    print(mexage)
                    #to delete
                    if months[i,0] == 12:
                        file_name_tmin = "D:\Andrea\SWB_Po\precip\\tmin_%04d.zip" % (year+1)
                        z_tmin = zipfile.ZipFile(file_name_tmin)
                        file_name_tmax = "D:\Andrea\SWB_Po\precip\\tmax_%04d.zip" % (year+1)
                        z_tmax = zipfile.ZipFile(file_name_tmax)
                        file_name_precip = "D:\Andrea\SWB_Po\precip\\precip_%04d.zip" % (year+1)
                        z_precip = zipfile.ZipFile(file_name_precip)
                    if i>1:
                        numberOfDays_old = calendar.monthrange(months[i-2,1], months[i-2,0])[1]
                        for day in  list(range(1, numberOfDays_old+1)):
                            day_file_tmin = "D:\Andrea\SWB_Po\precip\\tmin_%04d_%02d_%02d.asc" % (months[i-2,1],months[i-2,0],day)
                            os.remove(day_file_tmin)
                            day_file_tmax = "D:\Andrea\SWB_Po\precip\\tmax_%04d_%02d_%02d.asc" % (months[i-2,1],months[i-2,0],day)
                            os.remove(day_file_tmax)
                            day_file_precip = "D:\Andrea\SWB_Po\precip\\precip_%04d_%02d_%02d.asc" % (months[i-2,1],months[i-2,0],day)
                            os.remove(day_file_precip)                
    time.sleep(0.5)        
```
