The land cover map divides land into classes related to different land covers and land uses. 
To use the validated information, maps were loaded from the CORINE land cover inventory.

The data used to realize the map comes from the satellite imagery.Statistical classification procedures are based on spectral classification of a limited number of land cover classes, for detailed information on CLC, see Bossard et al 2000. The data can be downloaded after a login form the [Copernicus Land Monitoring service](https://land.copernicus.eu/pan-european/corine-land-cover). The dataset (map) is available every 6 year from the 2000. (2000; 2006; 2012; 2020).
The loaded data is than cutted and resampled with the same grid resolution used for the other data, using [QGIS](https://qgis.org/en/site/forusers/download.html). The final pixel resolution is 250*250 that is coarser than the original CLC resolution. The upscaling function used is the default QGIS rasrter upscaling function used in the export procedure (during the exporting procedure select the output pixel resolution and QGIS automatically rescale the raster). The function assigns the modal category of the fine pixels contained in the coarse (upscaled) pixel.
